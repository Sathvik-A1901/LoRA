```python
import torch
import numpy as np
_=torch.manual_seed(0)
```

```python
d,k=10,20

W_rank=2
W=torch.randn(d,W_rank)@torch.randn(W_rank,k)
print(W)
```

```python
W_rank=np.linalg.matrix_rank(W)
print(W_rank)
```

```python
U,S,V = torch.svd(W)

U_r=U[:, :W_rank]
s_r=torch.diag(S[:W_rank])
v_r=V[:, :W_rank].t()

S=U_r@s_r
A=v_r
print(S.shape)
print(A.shape)
```
